```python
from matplotlib import pyplot as plt
from skimage.feature import corner_harris, corner_peaks
from skimage.color import rgb2gray
```

```python
image = plt.imread('images/corners_building.jpg')
```

```python
plt.imshow(image)
```

```python
image_gray = rgb2gray(image)
```

```python
plt.imshow(image_gray)
```

```python
measure_image = corner_harris(image_gray)
```

```python
coords = corner_peaks(corner_harris(image_gray), min_distance=2)
coords_40 = corner_peaks(corner_harris(image_gray), min_distance=40)
```

```python
def show_image(image, title="Image", cmap_type='gray'):
    plt.imshow(image, cmap=cmap_type)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
show_image(image, "Original")
```

```python
def show_image_with_corners(image, coords, title="Corners detected"):
    plt.imshow(image, interpolation='nearest', cmap='gray')
    plt.title(title)
    plt.plot(coords[:, 1], coords[:, 0], '+r', markersize=15)
    plt.axis('off')
    plt.show()
```

```python
show_image_with_corners(image, coords)
```

```python
show_image_with_corners(image, coords_40)
```
